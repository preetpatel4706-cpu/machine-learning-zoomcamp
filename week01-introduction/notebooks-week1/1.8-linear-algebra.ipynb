# Multiplication

Vector Vector Multiplication

```python
def vector_vector_multiplication(u, v):
    assert u.shape == v.shape                                    # Vector-vector multiplication is only defined for vectors of the same shape
    n = u.shape[0]
    result = 0
    for i in range(n):
        result = result + u[i] * v[i]
    return result        
```

```python
vector_vector_multiplication(u, v)
```

```python
u.dot(v)       # Dot product of two vectors using numpy's built-in function
```

Matrix Vector Multiplication

```python
def matrix_vector_multiplication(U,v):
    assert U.shape[1] == v.shape[0]                              # Matrix-vector multiplication is only defined for matrices and vectors of compatible shapes
    num_rows = U.shape[0]
    result = np.zeros(num_rows)
    for i in range(num_rows):
        result[i] = vector_vector_multiplication(U[i], v)      # Multiply each row of the matrix by the vector
    return result
    
```

```python
matrix_vector_multiplication(U,v)
```

```python
U.dot(v)
```

Matrix Matrix Multiplication

```python
U = np.array([
    [1, 2],
    [3, 4]
])                 
```

```python
U.shape
```

```python
V = np.array([
    [5, 6],
    [7, 8]
])
```

```python
V.shape
```

```python
def matrix_matrix_multiplication(U, v):
    assert U.shape[1] == V.shape[0]                              # Matrix-matrix multiplication is only defined for matrices of compatible shapes
    num_rows = U.shape[0]
    num_cols = V.shape[1]
    result = np.zeros((num_rows, num_cols))
    for i in range(num_cols):
        vi = V[:, i]                                                # Get the i-th column of the second matrix
        Uvi = matrix_vector_multiplication(U, vi)                       # Multiply the first matrix by the i-th column of the second matrix
        result[:, i] = Uvi                                         # Store the result in the i-th column of the result matrix
    return result
```

```python
matrix_matrix_multiplication(   U, V)
```

```python
U.dot(V)
```
